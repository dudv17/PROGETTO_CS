# 3. For each (primary) breed, determine the ratio between the number of dogs of Mixed Breed and those not of Mixed Breed. Hint: look at the secondary_breed.

^ We now know how many occurences there are for each primary breed that haven't mixed with other breeds.

```python
breed_ratio = pd.merge(mixed_breed, not_mixed_breed, on='breed_primary')
breed_ratio
```

```python
breed_ratio['ratio'] = breed_ratio['mixed'] / breed_ratio['not_mixed']
breed_ratio[['breed_primary', 'ratio']]
```

# 5. For each state, compute the sex imbalance, that is the difference between male and female dogs. In which state this imbalance is largest?

```python
sex_counts = dogs.groupby(by='contact_state')['sex'].value_counts().unstack()
sex_counts
```

**dogs.groupby(by='contact_state')['sex'].value_counts() --** we group by state and count the number of occurences of each sex for each state  
**.unstack() --** we reshape the dataframe to have the values of 'sex' as columns and the states as index

```python
sex_counts['sex_imbalance'] = sex_counts['Male'] - sex_counts['Female']
sex_counts
```

```python
max_imbalance_state = sex_counts['sex_imbalance'].idxmax()
print(f'The state with the largest sex imbalance is: {max_imbalance_state}')
```

# 12. Find for duplicates in the dogs dataset. Two records are duplicates if they have (1) same breeds and sex, and (2) they share at least 90% of the words in the description field. Extra points if you find and implement a more refined for determining if two rows are duplicates.

```python
import re
```

```python
# Drop any rows with missing values
dogs_clean = dogs.dropna(subset=['breed_primary', 'description', 'sex'])
#dogs_clean
```

```python
# Preprocess the description field
def preprocess_description(desc):
    if pd.isna(desc):
        return []
    desc = re.sub(r'\W+', ' ', desc.lower())
    return desc.split()
```

^ this function takes a string as input and returns a list of preprocessed words. The preprocessing steps include converting the string to lowercase, removing any non-alphanumeric characters, and splitting the string into a list of words.

```python
# Define a function to calculate the percentage of shared words between two descriptions
def shared_word_percentage(desc1, desc2):
    set1 = set(desc1)
    set2 = set(desc2)
    if len(set1) == 0 or len(set2) == 0:
        return 0.0
    intersection = set1.intersection(set2)
    union = set1.union(set2)
    return len(intersection) / len(union)
```

^ this function is defined to calculate the percentage of shared words between two descriptions. This is done by first converting the descriptions to sets of words, and then calculating the intersection and union of the sets.

```python
# Preprocess the descriptions using a vectorized function
dogs_clean['description_processed'] = dogs_clean['description'].apply(preprocess_description)
```

^ the 'description_processed' column in the dogs_clean DataFrame is created by applying the 'preprocess_description' function to the 'description' column using the 'apply()' method.

```python
# Create a dictionary to store the processed descriptions and their corresponding IDs
desc_dict = dict(zip(dogs_clean['id'], dogs_clean['description_processed']))
#desc_dict
```

^ a dictionary called 'desc_dict' is created to store the processed descriptions and their corresponding IDs.

```python
# Use a hash table to store the processed descriptions and their corresponding IDs
desc_hash = {}
for i, desc in enumerate(dogs_clean['description_processed']):
    desc_hash[hash(' '.join(desc))] = desc_hash.get(hash(' '.join(desc)), []) + [dogs_clean.iloc[i]['id']]
```

```python
type(desc_hash)
```

^ a hash table called 'desc_hash' is created to store the processed descriptions and their corresponding IDs. The hash table is constructed by iterating over the 'description_processed' column in the dogs_clean DataFrame, and using the 'hash()' function to generate a hash value for each description. The hash value is used as a key in the desc_hash dictionary, and the corresponding ID is added to a list of IDs associated with that hash value.

```python
# Find the duplicate pairs using the hash table
duplicate_pairs = []
for i, desc in enumerate(dogs_clean['description_processed']):
    desc_hash[hash(' '.join(desc))].remove(dogs_clean.iloc[i]['id'])
    for j in desc_hash[hash(' '.join(desc))]:
        if dogs_clean.iloc[i]['breed_primary'] == dogs_clean[dogs_clean['id'] == j]['breed_primary'].values[0] and \
           dogs_clean.iloc[i]['breed_secondary'] == dogs_clean[dogs_clean['id'] == j]['breed_secondary'].values[0] and \
           dogs_clean.iloc[i]['sex'] == dogs_clean[dogs_clean['id'] == j]['sex'].values[0]:
            shared_words = shared_word_percentage(desc, desc_dict[j])
            if shared_words >= 0.9:
                duplicate_pairs.append((dogs_clean.iloc[i]['id'], j))
    desc_hash[hash(' '.join(desc))].append(dogs_clean.iloc[i]['id'])
```

^ a loop is used to iterate over the 'description_processed' column in the dogs_clean DataFrame. For each record, the ID is removed from the list of IDs associated with its hash value in the desc_hash dictionary. Then, a nested loop is used to compare the current record to all other records in the desc_hash dictionary with the same hash value. If the current record and another record have the same breeds and sex, and their shared word percentage is at least 90%, then they are considered a duplicate pair and their IDs are added to the duplicate_pairs list.

^^ execution time: ~ 1 minute
